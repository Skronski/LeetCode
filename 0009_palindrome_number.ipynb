# 9. Palindrome Number
**Example 1**:    
&emsp;**Input:** x = 121  
    &emsp;**Output:** true  
   &emsp;**Explanation:** 121 reads as 121 from left to right and from right to left.  
**Example 2:**  
   &emsp;**Input**: x = -121  
    &emsp;**Output**: false  
    &emsp;**Explanation**: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.  
**Example 3**:  
    &emsp;**Input:** x = 10  
    &emsp;**Output**: false  
    &emsp;**Explanation**: Reads 01 from right to left. Therefore it is not a palindrome.  


```python
def isPalindrome(x):
    """
    :type x: int
    :rtype: bool
    """
    x = str(x)
    for i in range(0,int(len(x)/2) + 1):
        if x[i] != x[len(x)-i-1]:
            return False
        return True
```
